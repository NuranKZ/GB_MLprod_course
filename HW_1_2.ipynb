#### Согласно принципу SMART, попробуйте расписать кейс кредитного скоринга, разбиравшийся на одном из предыдущих курсов

**Общая задача кредитного скоринга** - предсказание дефолта по переодичным платежам (процентные выплаты плюс (опционально) выплата части основного долга) клиентов

**Решение задачи в разрезе принципов SMART**
- Specific: Решаем задачу бинарной классификации c применением различных классификационных алгоритмов
- Measurable: Target: класс 1 - дефолт, класс 0 - не дефолт
- Attainable: Этапы построения модели:
  - Сбор исторических данных (БД)
  - Чистка, подготовка и преобразование данных в удобный для моделирования формат (SQL-request, PANDAS)
  - Разведочный анализ данных (корелляционный анализ, кластеризация и т.д., понижение размерности)
  - Feature Engeneering (подготовка итоговых данных для моделирования, разделение данных на тренировочную, валидационную и тестовую)
  - Поиск гиперпараметров моделей
  - Моделирование и финальная валидация
  - Подготовка прототипа для AB-тестирования
  - Внедрение модели в производственные системы 
- Relevant: Тестирование настроенного прототипа (AB-testing), расширение сбора данных (назад к шагу 3)
- Time-based: план-график проекта (Сбор данных - Разработка прототипа - Тестирование - Внедрение в производство - Механизмы обновления модели)

  - цены (тарифы) на аренду в разрезе арендуемых типов машин

  - техническое состояние арендуемых машин

  - дефицит арендуемых машин
  - периоды ожидания предоставления машин с момента запроса

  - причина ухода клиента (анкетирование)
  - имеется ли собственный автотранспорт у клиента
  - семейное положение
  - количество детей клиента
  - является ли клиент - командированным
  - направлен ли клиент в регион на ограниченное время по работе (secondment)
  - уголовное прошлое клиента (это тоже м.б. важным :)
  - недвижимость клиента (собственное, арендованное, отсутствует - живет в отелях)
  - нагрузка по кредитным платежам клиента

#### Напишите функцию, которая возвращает сумму двух вещественных аргументов a и b, а к ней декоратор, который делает так, чтобы возвращаемое значение функцией было по модулю 5.

```python
# 1. Декоратор и функция
def decor_mod_5(f):
    def wrapper(*args, **kwargs):
        value = f(*args, **kwargs)
        return value // 5
    return wrapper

@ decor_mod_5
def f(a,b):
    return a+b

# 2. Check
print(f(10, 16))
print((10+16)//5)
```

#### Попробуйте описать своими словами основные отличия SMOTE от ADASYN, ознакомившись с документацией к ним

**SMOTE** - данный метод для oversampling работает следующим образом
- для каждой точки минорного класса запускает алгоритм нахождения n-ближайших соседей, тем самым определяя области пространства как кластеры минорных классов с соответствующими центрами
- проводит прямые линии между центрами кластеров
- генерирует синтетические (новые) точки минорных классов на этих линиях

**ADASYN**
- в целом алгоритм генерации синтетических точек похож
- основное отличие в том, что при генерации синтетических точек к ним (по пространству Х) добавляется некоторый небольшой случайный шум, что делает точки несколько более реалистичными
